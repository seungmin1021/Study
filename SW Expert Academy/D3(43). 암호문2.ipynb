```python
for i in range(1):
    n = int(input())
    secret = list(map(str,input().split()))
    com_num = int(input())
    command = list(map(str, input().split()))

    for j in range(com_num):
        if command[0] == 'I':
            temp = command[:3+int(command[2])]
            command = command[3+int(command[2]):]
            secret = secret[:int(temp[1])] + temp[3:] + secret[int(temp[1]):]
        elif command[0] == 'D':
            temp = command[:3]
            command = command[3:]
            secret = secret[:int(temp[1])] + secret[int(temp[1])+int(temp[2]):]
    print('#{}'.format(i+1),end=' ')
    for j in range(10):
        print(secret[j],end=' ')
    print()
```
