```python
for i in range(10):
    n = int(input())
    a = list(map(int,input().split()))
    leng = len(a)
    for j in range(2):
        a.append(0)
    s = 0
    for j in range(2,leng):
        if a[j] > max(a[j-2],a[j-1],a[j+1],a[j+2]):
            s += a[j] - max(a[j-2],a[j-1],a[j+1],a[j+2])
    print('#{} {}'.format(i+1, s))
```
