```python
for i in range(1):
    n = int(input())
    a = [list(map(int, input().split())) for _ in range(n)]
    cnt = 0
    for j in range(n):
        f = 0
        for k in range(n):
            if a[k][j] == 1:
                f = 1
            elif a[k][j] == 2 and f == 1:
                cnt += 1
                f = 0
    print('#{} {}'.format(i+1, cnt))
```
