```python
t = int(input())
for i in range(t):
    n, m = map(int,input().split())
    result = 0
    if n == m:
        print('#{} {} {}'.format(i+1, n, result))
    else:
        while True:
            n -= 2
            m -= 1
            result += 1
            if n == m:
                print('#{} {} {}'.format(i+1, n, result))
                break
```
