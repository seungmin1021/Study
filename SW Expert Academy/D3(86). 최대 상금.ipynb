```python
def dfs(n):
    global answer
    if not n:
        temp = int(''.join(s))
        if answer < temp:
            answer = temp
        return
        
    for i in range(leng):
        for j in range(i+1, leng):
            s[i], s[j] = s[j], s[i]
            temp = ''.join(s)
            if visited.get((temp, n-1), 1):
                visited[(temp, n-1)] = 0
                dfs(n-1)
            s[i], s[j] = s[j], s[i]

for t in range(int(input())):
    answer = 0
    s, n = input().split()
    s = list(s)
    n = int(n)
    leng = len(s)
    visited = {}
    dfs(n)
    print('#{} {}'.format(t+1, answer))
```
