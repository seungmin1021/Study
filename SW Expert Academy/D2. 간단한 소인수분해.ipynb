```python
n = int(input())
k = [2,3,5,7,11]
ans = [0] * len(k)
for i in range(n):
    num = int(input())
    while num > 1:
        for j in k:
            if num % j == 0:
                num = num // j
                ans[k.index(j)] += 1
    print('#{}'.format(i+1),end=' ')
    for j in ans:
        print(j,end=' ')
    print()
    ans = [0] * len(k)
```
