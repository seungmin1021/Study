```python
t = int(input())
for i in range(t):
    n, q = map(int,input().split())
    result = ['0' for _ in range(n)]
    a = [list(map(int,input().split())) for _ in range(q)]
    for j in range(q):
        for k in range(a[j][0]-1,a[j][1]):
            result[k] = str(j+1)
    print('#{} {}'.format(i+1, ' '.join(result)))
```
