```python
import math
t = int(input())
score = [0,'A+','A0','A-','B+','B0','B-','C+','C0','C-','D0']
for i in range(t):
    n, target = map(int,input().split())
    a = []
    for j in range(n):
        m, f, s = map(int,input().split())
        a.append((j+1,(m*0.35)+(f*0.45)+(s*0.2)))
    a.sort(key=lambda x : x[1],reverse = True)
    idx = 0
    for j in range(a):
        if a[j][0] == target:
            idx = j+1
    print('#{} {}'.format(i+1, score[math.ceil(idx /(n//10))]))
```
