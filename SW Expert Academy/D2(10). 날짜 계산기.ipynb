```python
t = int(input())
dic = {1:31,2:28,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}
for i in range(t):
    a = list(map(int,input().split()))
    sum = 0
    for j in range(a[0],a[2]+1):
        if j == a[0]:
            sum += dic[j]-a[1]
        elif j == a[2]:
            sum += a[3]
        else:
            sum += dic[j]
    print('#{} {}'.format(i+1,sum+1))
```
