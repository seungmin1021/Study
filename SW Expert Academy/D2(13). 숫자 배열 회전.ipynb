```python
t = int(input())
for i in range(t):
    n = int(input())
    nums = [list(map(int,input().split())) for i in range(n)]
    print('#{}'.format(i+1))
    result = [[0 for _ in range(3)] for _ in range(3)]
    for j in range(n):
        s = ''
        for k in range(n-1,-1,-1):
            s += str(nums[k][j])
        result[j][0] = s
    for j in range(n-1,-1,-1):
        s = ''
        for k in range(n-1,-1,-1):
            s += str(nums[j][k])
        result[n-j-1][1] = s
    for j in range(n-1,-1,-1):
        s = ''
        for k in range(n):
            s += str(nums[k][j])
        result[n-j-1][2] = s
    for i in result:
        print(' '.join(i))
```
