```python
for i in range(10):
    n = int(input())
    a = [list(input()) for _ in range(8)]
    result = 0
    for j in range(8):
        for k in range(8-n+1):
            s = a[j][k:k+n]
            if s == s[::-1]:
                result += 1
    for j in range(8):
        for k in range(8-n+1):
            s = ''
            for x in range(n):
                s += a[k+x][j]
            if s == s[::-1]:
                result += 1
    print('#{} {}'.format(i+1, result))
```
