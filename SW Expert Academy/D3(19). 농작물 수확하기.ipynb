```python
t = int(input())
for i in range(t):
    n = int(input())
    a = [list(input()) for _ in range(n)]
    l = 0
    r = n
    sum = 0
    for j in range((n//2), -1,-1):
        for k in range(l,r):
            sum += int(a[j][k])
        l+=1
        r-=1
    l = 1
    r = n-1
    for j in range(n//2+1,n):
        for k in range(l,r):
            sum += int(a[j][k])
        l+=1
        r-=1
    print('#{} {}'.format(i+1, sum))
```
