```python
for i in range(1):
    n = int(input())
    a = list(map(int,input().split()))
    while a[-1]>0:
        for j in range(5):
            x = a.pop(0)
            if x-(j+1)<=0:
                a.append(0)
                break 
            else:
                a.append(x-(j+1))

    print('#{}'.format(n),end=' ')
    for j in a:
        print(j,end=' ')
    print()
```
