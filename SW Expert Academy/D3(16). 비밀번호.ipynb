```python
for i in range(1):
    n, s = map(str, input().split())
    s = list(s)
    a = []
    for j in s:
        if len(a) == 0:
            a.append(j)
            continue
        if a[-1] == j:
            a.pop()
        else:
            a.append(j)
    print('#{}'.format(i+1),end=' ')
    for j in a:
        print(j,end='')
    print()
```
