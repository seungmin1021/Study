```python
a = int(input())
for i in range(1,a+1):
    n, k = map(int,input().split())
    cnt = 0
    board = [list(map(str,input().split())) for i in range(n)]
    for j in board:
        cnt += ''.join(j).split('0').count('1'*k)
    for j in zip(*board):
        cnt += ''.join(j).split('0').count('1'*k)
    print('#{} {}'.format(i,cnt))
```
