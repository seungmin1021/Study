```python
t = int(input())
for i in range(t):
    hap = []
    a, b = map(int,input().split())
    al = list(map(int,input().split()))
    bl = list(map(int,input().split()))
    if b < a:
        al, bl = bl, al
    for k in range(abs(a-b)+1):
        sum = 0
        for j in range(len(al)):
            sum += al[j]*bl[j]
        hap.append(sum)
        bl.pop(0)
    
    print('#{} {}'.format(i+1,max(hap)))
```
