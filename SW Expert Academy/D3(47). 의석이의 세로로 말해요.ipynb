```python
t = int(input())
for i in range(t):
    a = [input() for _ in range(5)]
    x = max(len(a[0]),len(a[1]),len(a[2]),len(a[3]),len(a[4]))
    b = []
    for j in range(5):
        a[j] = a[j].ljust(x,' ')
        b.append(list(a[j]))
    print(b)
    result = ''
    for j in range(x):
        for k in range(5):
            if b[k][j] == ' ':
                continue
            else:
                result += b[k][j]
    print('#{} {}'.format(i+1, result))
```
