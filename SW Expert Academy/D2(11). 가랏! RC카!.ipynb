```python
t = int(input())
for i in range(t):
    tt = int(input())
    m = 0
    speed = 0
    for j in range(tt):
        c = list(map(int,input().split()))
        if c[0] == 1:
            speed += c[1]
        elif c[0] == 2:
            speed -= c[1]
        if speed < 0:
            speed = 0
        m += speed
    print('#{} {}'.format(i+1, m))
```
