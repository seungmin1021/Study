```python
for t in range(int(input())):
    sec = {'0001101':0, '0011001':1, '0010011':2, '0111101':3, '0100011':4,
           '0110001':5, '0101111':6, '0111011':7, '0110111':8, '0001011':9}
    n, m = map(int,input().split())
    secret = [input() for _ in range(n)]
    s = ''
    for i in secret:
        if '1' not in i:
            continue
        else:
            for j in range(m-1,-1,-1):
                if i[j] == '1':
                    s += i[j-55:j+1]
                    break
            break
    r = []
    for i in range(0,56,7):
        r.append(sec[s[i:i+7]])
    x = ((r[0]+r[2]+r[4]+r[6])*3) + (r[1]+r[3]+r[5]+r[7])
    if x % 10 == 0:
        print('#{} {}'.format(t+1, sum(r)))
    else:
        print('#{} {}'.format(t+1, 0))
```
