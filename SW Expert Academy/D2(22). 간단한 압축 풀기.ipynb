```python
t = int(input())
for i in range(t):
    n = int(input())
    a = []
    for j in range(n):
        c, k = map(str,input().split())
        for k in range(int(k)):
            a.append(c)
    print('#{}'.format(i+1))
    for j in range(len(a)):
        if j>0 and j % 10 == 0:
            print()
        print(a[j],end='')
    print()
```
