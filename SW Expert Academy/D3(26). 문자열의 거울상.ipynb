```python
t = int(input())
for i in range(t):
    s = input()
    result = ''
    for j in s[::-1]:
        if j == 'b':
            result += 'd'
        elif j == 'd':
            result += 'b'
        elif j == 'p':
            result += 'q'
        else:
            result += 'p'
    print('#{} {}'.format(i+1, result))
```
