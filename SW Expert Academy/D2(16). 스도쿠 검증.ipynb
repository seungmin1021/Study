```python
t = int(input())
for i in range(t):
    s = [list(map(int,input().split())) for _ in range(9)]
    f = 1
    for x in range(0,9,3):
        for y in range(0,9,3):
            sum = 0
            for j in range(3):
                for k in range(3):
                    sum += s[j+x][k+y]
            if sum != 45:
                f = 0
    for j in range(9):
        sum = 0
        for k in range(9):
            sum += s[j][k]
        if sum != 45:
            f = 0

    for j in range(9):
        sum = 0
        for k in range(9):
            sum += s[k][j]
        if sum != 45:
            f = 0
    print('#{} {}'.format(i+1, f))
```
