```python
y = {'01':31,'02':28,'03':31,'04':30,'05':31,'06':30,'07':31,'08':31,'09':30,'10':31,'11':30,'12':31}
n = int(input())
for i in range(n):
    st = input()
    ans = ''
    ans += st[:4]+'/'
    if st[4:6] in y:
        ans += st[4:6]+'/'
    else:
        print('#{} -1'.format(i+1))
        continue
    if int(st[6:]) > 0 and int(st[6:]) <= y[st[4:6]]:
        ans += st[6:]
    else:
        print('#{} -1'.format(i+1))
        continue
    print('#{} {}'.format(i+1,ans))
```
