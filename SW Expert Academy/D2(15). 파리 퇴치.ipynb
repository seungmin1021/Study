```python
t = int(input())
for i in range(t):
    n, m= map(int, input().split())
    a = [list(map(int,input().split())) for _ in range(n)]
    result = []
    for x in range(n-m+1):
        for y in range(n-m+1):
            sum = 0
            for j in range(m):
                for k in range(m):
                    sum += a[x+j][k+y]
            result.append(sum)
    print('#{} {}'.format(i+1, max(result)))
```
