```python
for i in range(10):
    n = int(input())
    a = [list(map(int,input().split())) for _ in range(100)]
    result = []
    s = 0
    for j in range(100):
        s = sum(a[j])
        result.append(s)
    for j in range(100):
        s = 0
        for k in range(100):
            s += a[j][k]
        result.append(s)
    s = 0
    for j in range(100):
        s += a[j][j]
        result.append(s)
    s = 0
    for j in range(100):
        s += a[99-j][99-j]
    print('#{} {}'.format(n, max(result)))
```
