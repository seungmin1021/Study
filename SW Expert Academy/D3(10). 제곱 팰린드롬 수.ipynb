```python
from math import sqrt
t = int(input())
for i in range(t):
    a, b = map(int,input().split())
    cnt = 0
    for j in range(a,b+1):
        if str(j) == str(j)[::-1] and sqrt(j) == int(sqrt(j)): 
            if str(int(sqrt(j))) == str(int(sqrt(j)))[::-1]:
                cnt += 1
    print('#{} {}'.format(i+1, cnt))
```
