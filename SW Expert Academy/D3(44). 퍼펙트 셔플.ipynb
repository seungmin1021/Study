```python
t = int(input())
for i in range(t):
    n = int(input())
    a = list(map(str, input().split()))
    b = a[:(n//2)+(n%2)]
    c = a[(n//2)+(n%2):]
    result = []
    for j, k in zip(b,c):
        result.append(j)
        result.append(k)
    if len(result) != n:
        result.append(b[-1])
    print('#{} {}'.format(i+1, ' '.join(result)))
```
