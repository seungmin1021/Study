```python
def check(arr):
    for k in range(100):
        for y in range(100):
            for x in range(0,k+1):
                s = arr[y][x:100-k+x]
                if s == s[::-1]:
                    return len(s)
                        
for i in range(10):
    n = int(input())
    a = [list(input()) for _ in range(100)]
    b = list(zip(*a))
    print('#{} {}'.format(i+1, max(check(a),check(b))))
```
