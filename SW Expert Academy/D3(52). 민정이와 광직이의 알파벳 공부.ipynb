```python
from itertools import combinations as c
t = int(input())
for i in range(t):
    n = int(input())
    a = [input() for _ in range(n)]
    cnt = 0
    for j in range(n):
        b = list(c(a,j+1))
        for k in b:
            s = ''
            for x in k:
                s += x
            for x in range(ord('a'), ord('z')+1):
                if chr(x) not in s:
                    break
            else:
                cnt +=1
    print('#{} {}'.format(i+1, cnt))
```
