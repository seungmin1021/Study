```python
def solution(lottos, win_nums):
    dic = {0:6, 1:6, 2:5, 3:4, 4:3, 5:2, 6:1}
    answer = []
    cnt = 0
    zero = lottos.count(0)
    for i in lottos:
        if i in win_nums:
            cnt += 1
    answer.append(dic[cnt+zero])
    answer.append(dic[cnt])

    return answer
```

```python
solution([0, 0, 0, 0, 0, 0], [38, 19, 20, 40, 15, 25])
```
