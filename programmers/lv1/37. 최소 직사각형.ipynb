```python
def solution(sizes):
    answer = 0
    for i in sizes:
        if i[0] < i[1]:
            i[0], i[1] = i[1], i[0]
    sizes.sort(key = lambda x : x[0],reverse = True)
    print(sizes)
    a = sizes[0][0]
    sizes.sort(key = lambda x : x[1], reverse = True)
    print(sizes)
    b = sizes[0][1]
    answer = a * b
    return answer
```
