```python
def solution(id_list, report, k):
    attemp = dict.fromkeys(id_list,'')
    reported = dict.fromkeys(id_list,0)
    mail = dict.fromkeys(id_list,0)

    for i in set(report):
        x, y = i.split(' ')
        reported[y] += 1
        attemp[x] += y

    print(attemp, reported)
    for i in reported:
        if reported[i] >= k:
            for j in attemp:
                if i in attemp[j]:
                    mail[j] += 1
    print(mail)
    return list(mail.values())
```

```python
def solution(id_list, report, k):
    answer = [0] * len(id_list)
    reported = dict.fromkeys(id_list,0)

    for i in set(report):
        x, y = i.split(' ')
        reported[y] += 1

    print(reported)
    for r in set(report):
        x,y = r.split()
        if reported[y] >= k:
            answer[id_list.index(x)] += 1
    return answer
```
