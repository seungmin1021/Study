```python
def solution(survey, choices):
    answer = ''
    sur = {1:3, 2:2, 3:1, 4:0, 5:1, 6:2, 7:3}
    score = {'R':0, 'T':0, 'C':0, 'F':0, 'J':0, 'M':0, 'A':0, 'N':0}

    for i in range(len(survey)):
        if choices[i] < 4:
            score[survey[i][0]] += sur[choices[i]]
        elif choices[i] > 4:
            score[survey[i][1]] += sur[choices[i]]
        else:
            continue

    a = list(score.items())
    
    for i in range(0,7,2):
        if a[i][1] > a[i+1][1] or a[i][1] == a[i+1][1]:
            answer += a[i][0]
        elif a[i][1] < a[i+1][1]:
            answer += a[i+1][0]

    return answer
```

```python
solution(["AN", "CF", "MJ", "RT", "NA"],[5, 3, 2, 7, 5])
```
