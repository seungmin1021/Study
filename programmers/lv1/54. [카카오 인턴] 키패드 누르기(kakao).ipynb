```python
import math
def distance(x1,y1,x2,y2):
    x = abs(x2-x1)
    y = abs(y2-y1)
    xy = x + y
    return xy
def solution(numbers, hand):
    answer = ''
    key = {1:[0,0], 2:[0,1], 3:[0,2],
           4:[1,0], 5:[1,1], 6:[1,2],
           7:[2,0], 8:[2,1], 9:[2,2],
           '*':[3,0], 0:[3,1], '#':[3,2]}
    l = key['*']
    r = key['#']
    l_c = [1,4,7]
    r_c = [3,6,9]
    for i in numbers:
        if i in l_c:
            l = key[i]
            answer += 'L'
        elif i in r_c:
            r = key[i]
            answer += 'R'
        else:
            if distance(r[0], r[1], key[i][0], key[i][1]) == distance(l[0], l[1], key[i][0], key[i][1]):
                if hand == 'right':
                    answer += 'R'
                    r = key[i]
                else:
                    answer += 'L'
                    l = key[i]
            elif distance(r[0], r[1], key[i][0], key[i][1]) < distance(l[0], l[1], key[i][0], key[i][1]):
                answer += 'R'
                r = key[i]
            elif distance(r[0], r[1], key[i][0], key[i][1]) > distance(l[0], l[1], key[i][0], key[i][1]):
                answer += 'L'
                l = key[i]
        
    return answer
```
