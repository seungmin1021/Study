```python
def solution(a):
    re = []
    cnt = 0
    for i in range(len(a)):
        if a[i].isdigit():
            if a[i-1].isdigit():
                re[cnt-1] += 9
            else:
                re.append(int(a[i]))
                cnt += 1
        elif a[i].isalpha():
            if a[i] == 'D':
                re[cnt-1] = re[cnt-1] * re[cnt-1]
            elif a[i] == 'T':
                re[cnt-1] = (re[cnt-1] ** 2) * re[cnt-1]
        elif a[i] == '*':
            if cnt == 1:
                re[cnt-1] = re[cnt-1] * 2
            elif cnt > 1:
                re[cnt-1] = re[cnt-1] * 2
                re[cnt-2] = re[cnt-2] * 2
        elif a[i] == '#':
            re[cnt-1] = re[cnt-1] * -1

    return sum(re)
```

```python
solution('1D2S#10S')
```
