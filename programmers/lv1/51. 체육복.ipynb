```python
def solution(n, lost, reserve):
    a = []
    for i in lost:
        if i in reserve:
            a.append(i)

    for i in a:
        reserve.pop(reserve.index(i))
        lost.pop(lost.index(i))
        
    reserve.sort()
    lost.sort()

    for i in range(len(lost)):
        if lost[0]-1 in reserve:
            reserve.pop(reserve.index(lost[0]-1))
            lost.pop(0)
            continue
        elif lost[0]+1 in reserve:
            reserve.pop(reserve.index(lost[0]+1))
            lost.pop(0)
            continue
        lost.append(lost.pop(0))

    return n-len(lost)
```

```python
solution(5,[2,4],[2,3,5])
```
