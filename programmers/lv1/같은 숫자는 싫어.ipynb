```python
def solution(arr):
    answer = []
    for i in arr:
        while answer and answer[-1] == i:
            answer.pop()
        answer.append(i)
            
    return answer
```

```python
def solution(arr):
    answer = []
    for i in arr:
        if len(answer) == 0 or answer[-1] != i:
            answer.append(i)
            
    return answer
```
