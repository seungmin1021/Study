```python
def solution(s, n):
    answer = ''
    for i in s:
        if i.isupper():
            if ord(i)+n >90:
                a = abs(ord(i)+n-90)
                answer += chr(64+a)
            else:
                answer += chr(ord(i)+n)
        elif i.islower():
            if ord(i)+n >122:
                a = abs(ord(i)+n-122)
                answer+= chr(96+a)
            else:
                answer += chr(ord(i)+n)
        elif i == ' ':
            answer += ' '
            
    return answer
```
