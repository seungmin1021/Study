```python
def solution(k, score):
    answer = []
    result = []
    for i in score:
        if len(answer) < k:
            answer.append(i)
        else:
            if i > min(answer):
                answer.pop(answer.index(min(answer)))
                answer.append(i)
        result.append(min(answer))
    return result
```

```python
solution(3,[10, 100, 20, 150, 1, 100, 200])
```
