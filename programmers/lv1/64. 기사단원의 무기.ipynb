```python
def solution(number, limit, power):
    att = []
    for i in range(1,number+1):
        cnt = 0
        for j in range(1,int(i**(1/2))+1):
            if i % j == 0:
                cnt += 1
                if j ** 2 != i:
                    cnt += 1
        att.append(cnt)
    for i in range(number):
        if att[i] > limit:
            att[i] = power
    
    return sum(att)
```

```python
solution(5,3,2)
```
