```python
def solution(answers):
    answer = []
    
    one = [1,2,3,4,5]
    two = [2,1,2,3,2,4,2,5]
    three = [3,3,1,1,2,2,4,4,5,5]
    
    point1 = 0
    point2 = 0
    point3 = 0
    
    for i in range(len(answers)):
        if(answers[i] == one[i%5]):
            point1+=1
        if(answers[i] == two[i%8]):
            point2+=1
        if(answers[i] == three[i%10]):
            point3+=1
            
    k = max(point1, point2, point3)
        
    if(k==point1):
        answer.append(1)
    if(k==point2):
        answer.append(2)
    if(k==point3):
        answer.append(3)
        
    return answer
```
