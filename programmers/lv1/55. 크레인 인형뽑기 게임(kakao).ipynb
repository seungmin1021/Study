```python
def solution(board, moves):
    answer = 0
    bas = []
    for i in range(len(moves)):
        for j in range(len(board)):
            if board[j][moves[i]-1] != 0:
                bas.append(board[j][moves[i]-1])
                board[j][moves[i]-1] = 0
                break
        if len(bas) > 1 and bas[-1] == bas[-2]:
            bas.pop()
            bas.pop()
            answer += 2
    return answer
```

```python
board = [[0,0,0,0,0],
        [0,0,1,0,3],
        [0,2,5,0,1],
        [4,2,4,4,2],
        [3,5,1,3,1]]
moves = [1,5,3,5,1,2,1,4]

solution(board, moves)
```
