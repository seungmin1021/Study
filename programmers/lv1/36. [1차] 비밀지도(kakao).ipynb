```python
def change(i,n):
    a = []
    while n >= 1:
        m = n % 2
        a.append(str(m))
        n = n // 2
        if n == 1:
            a.append(str(n))
            break
    while len(a) < i:
        a.append('0')
    return ''.join(a[::-1])

def solution(n, arr1, arr2):
    answer = []
    one = []
    two = []
    for i in arr1:
        one.append(change(n,i))
    for i in arr2:
        two.append(change(n,i))
    ostr = ''
    tstr = ''
    toNS1 = []
    toNS2 = []
    for i in one:
        for j in range(len(i)):
            if i[j] == '1':
                ostr += '#'
            else:
                ostr += ' '
        toNS1.append(ostr)
        ostr = ''
        
    for i in two:
        for j in range(len(i)):
            if i[j] == '1':
                tstr += '#'
            else:
                tstr += ' '
        toNS2.append(tstr)
        tstr = ''
    rstr = ''
    rNS = []
    for i in range(n):
        for j in range(n):
            if toNS1[i][j] == '#' or toNS2[i][j] == '#':
                rstr += '#'
            elif toNS1[i][j] == ' ' and toNS2[i][j] == ' ':
                rstr += ' '
        rNS.append(rstr)
        rstr = ''
        
    return rNS
```
