```python
def solution(n, m):
    answer = []
    a = []
    if n > m:
        n, m = m, n
    for i in range(n):
        if n % (i+1) == 0 and m % (i+1) == 0:
            a.append(i+1)
    answer.append(max(a))
    
    for i in range(n):
        if (m*(i+1)) % n == 0:
            answer.append(m*(i+1))
            return answer
```
