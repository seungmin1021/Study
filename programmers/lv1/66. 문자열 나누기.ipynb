```python
def solution(s):
    answer = 0
    cnt = 0
    first = 0
    while len(s)>0:
        for i in range(len(s)):
            if s[i] == s[0]:
                first += 1
            else :
                cnt += 1

            if first == cnt or i == len(s)-1:
                s = s[i+1:]
                answer += 1
                cnt = 0
                first = 0
                break
    return answer
```

```python
solution('banana')
```
