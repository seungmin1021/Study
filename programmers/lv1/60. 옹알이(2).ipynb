```python
def solution(badding):
    answer = 0
    a = ['aya','ye','woo','ma']
    for i in badding:
        for j in a:
            if j*2 in i:break
            i = i.replace(j,' ')
        i = i.replace(' ','')
        if i == '':
            answer += 1
    return answer
```

```python
solution(["aya", "yee", "u"])
```
