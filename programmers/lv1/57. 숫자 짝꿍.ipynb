```python
def solution(x, y):
    answer = ''
    num = ['9','8','7','6','5','4','3','2','1','0']
    for i in num:
        answer += i * min(x.count(i),y.count(i))

    if len(answer) == 0:
        return '-1'
    elif answer[0] == '0':
        return '0'
    else:
        return answer
```
