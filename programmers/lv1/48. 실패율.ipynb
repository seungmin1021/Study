```python
def solution(N, stages):
    answer = []
    leng = len(stages)
    for i in range(N):
        cnt = 0
        for j in stages:
            if j == i+1:
                cnt += 1
        try:
            answer.append(cnt/leng)
            leng -= cnt
        except ZeroDivisionError:
            answer.append(0)
    result = [(i,v) for i, v in enumerate(answer)]
    result.sort(key = lambda x : x[1],reverse = True)

    return [result[i][0]+1 for i in range(len(result))]
```

```python
def solution(N, stages):
    answer = []
    dic = {}
    leng = len(stages)
    for i in range(N):
        cnt = 0
        for j in stages:
            if j == i+1:
                cnt += 1
        if cnt != 0:
            dic[i+1] = cnt/leng
            leng -= cnt
        else:
            dic[i+1] = 0

    dic2 = sorted(dic.items(), key = lambda x : x[1], reverse = True)
    for i in range(len(dic2)):
        answer.append(dic2[i][0])

    return answer
```
