```python
def solution(new_id):
    answer = ''
    #1단계
    for i in new_id:
        if i.isupper():
            answer += i.lower()
        else:
            answer += i
    #2단계
    a = ['-','_','.']
    answer2 = ''
    for i in answer:
        if i.isdigit() or i.isalpha() or i in a:
            answer2 += i
        else:
            continue

    #3단계
    answer3 = ''
    j = ''
    for i in answer2:
        if i == '.' and i == j:
            j = i
            continue
        j = i
        answer3 += i
    #4단계
    if len(answer3) > 0 and answer3[0] == '.':
        answer3 = answer3[1:]
    if len(answer3) > 0 and answer3[-1] == '.':
        answer3 = answer3[:-1]

    #5단계
    if answer3 == '':
        answer3 += 'a'

    #6단계
    if len(answer3) > 15:
        answer3 = answer3[:15]
    if answer3[-1] == '.':
        answer3 = answer3[:-1]

    #7단계
    if len(answer3) < 3:
        while len(answer3) < 3:
            answer3 += answer3[-1]
    return answer3
```

```python
solution("=.=")
```
