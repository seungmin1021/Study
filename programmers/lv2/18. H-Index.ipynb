```python
def solution(c):
    answer = 0
    h = 1
    a = []
    for i in range(max(c)):
        answer = 0
        for i in c:
            if i >= h:
                answer += 1
        if h <= answer:
            a.append(h)
        h += 1
    if len(a) == 0:
        return 0
    return a[-1]
```

```python
solution([6, 5, 5, 5, 3, 2, 1, 0]) 
```
