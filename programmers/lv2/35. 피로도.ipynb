```python
answer = 0
visited = []
def dfs(k, dun, cnt):
    global answer
    if cnt > answer:
        answer = cnt
    
    for i in range(len(dun)):
        if k >= dun[i][0] and not visited[i]:
            visited[i] = 1
            dfs(k-dun[i][1], dun, cnt+1)
            visited[i] = 0

def solution(k, dungeons):
    global visited
    visited = [0] * len(dungeons)
    dfs(k, dungeons, 0)
    return answer
```

```python
solution(80,[[80,20],[50,40],[30,10]])
```
