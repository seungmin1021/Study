```python
from math import factorial as f
def solution(x):
    answer = 0
    n = 1
    for i in range(x-1,(x//2+x%2)-1,-1):
        if x % 2 == 0:
            if i == x//2:
                answer += 1
            else:
                answer += f(i)//(f(n)*f(i-n))
                n+=1
        else:
            if i == x//2+x%2:
                answer += i
            else:
                answer += f(i)//(f(n)*f(i-n))
                n+=1
    return (answer + 1) % 1234567
```
