```python
def solution(n, words):
    answer = []
    cnt = 0
    stack = []
    for i in range(len(words)):
        if len(stack) == 0:
            stack.append(words[i])
            cnt += 1
            continue
        if words[i][0] == stack[-1][-1]:
            if words[i] in stack:
                cnt += 1
                if cnt % n == 0:
                    return [n, cnt//n]
                else:
                    return [cnt % n ,cnt//n+1]
            else:
                stack.append(words[i])
                cnt += 1
                
            if i == len(words)-1:
                return [0, 0]
        else:
            cnt += 1
            if cnt % n == 0:
                return [n, cnt//n]
            else:
                return [cnt % n ,cnt//n+1]
```
