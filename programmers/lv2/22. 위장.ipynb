```python
def solution(clothes):
    answer = 1
    dic = {}
    for i in clothes:
        dic[i[1]] = []
    for i in clothes:
        dic[i[1]].append(i[0])
    for i in dic:
        answer *= len(dic[i]) + 1
    return answer - 1
```

```python
solution([["yellow_hat", "headgear"], ["blue_sunglasses", "eyewear"], ["green_turban", "headgear"]])
```
