```python
def solution(s):
    if s[0] == ')'or s[-1] == '(':
        return False
    n = 0
    for i in range(0,len(s)):
        if s[i] == '(':
            n += 1
        else:
            n -= 1
        if n < 0:
            return False
    if n == 0:
        return True
    else:
        return False
```

```python
solution('())(()')
```
