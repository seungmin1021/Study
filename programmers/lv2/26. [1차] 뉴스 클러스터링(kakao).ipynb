```python
def strSplit(st):
    a = []
    for i in range(len(st)-1):
        if st[i:i+2].isalpha():
            a.append(st[i:i+2])
    return a

def solution(str1, str2):
    answer = 0
    s1 = strSplit(str1)
    s2 = strSplit(str2)
    
    lengS2 = len(s2)
    for i in s1:
        if i in s2:
            answer += 1
            s2.pop(s2.index(i))
    union = (len(s1)-answer) + (lengS2-answer) + answer

    if answer == 0 and union == 0:
        return 65536
    if answer == 0 and union > 0:
        return 0
    return int((answer / union) * 65536)
```

```python
solution('handshake','shake hands')
```
