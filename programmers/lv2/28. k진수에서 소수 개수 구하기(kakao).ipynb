```python
def solution(n, k):
    a = ''
    while n:
        a = str(n%k) + a
        n = n // k
    a = list(a.split('0'))

    cnt = 0
    for i in a:
        if i == '' or int(i) < 2:
            continue
        f = True
        for j in range(2,int(int(i)**0.5+1)):
            if int(i) % j == 0:
                f = False
                break
        if f:
            cnt += 1
    return cnt
```
