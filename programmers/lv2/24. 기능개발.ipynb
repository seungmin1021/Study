```python
def solution(progresses, speed):
    answer = []
    for i in range(len(progresses)):
        if (100 - progresses[i]) % speed[i] == 0:
            answer.append((100 - progresses[i]) // speed[i])
        else:
            answer.append((100 - progresses[i]) // speed[i] + 1)
    print(answer)
    a = []
    b = []
    cnt = 0
    for i in answer:
        if len(a) == 0:
            a.append(i)
            cnt += 1
            continue
        if a[-1] >= i:
            cnt += 1
        else:
            b.append(cnt)
            cnt = 0
            a.append(i)
            cnt += 1
    b.append(cnt)
    return b
```

```python
def solution(progresses, speeds):
    answer = []
    time = 0
    count = 0
    while len(progresses)> 0:
        if (progresses[0] + time*speeds[0]) >= 100:
            progresses.pop(0)
            speeds.pop(0)
            count += 1
        else:
            if count > 0:
                answer.append(count)
                count = 0
            time += 1
    answer.append(count)
    return answer
```
