```python
def solution(p, l):
    answer = 0
    q = [(i, v) for i, v in enumerate(p)]
    
    while True:
        cur = q.pop(0)
        if any(cur[1]<x[1] for x in q):
            q.append(cur)
        else:
            answer += 1
            if cur[0] == l:
                return answer
```

```python
solution([2,1,3,2],2)
```
