```python
from collections import deque
def solution(prices):
    answer = []
    queue = deque(prices)

    while queue:
        cur = queue.popleft()
        a = 0
        for i in queue:
            a += 1
            if cur > i:
                break
        answer.append(a)
    return answer
```

```python
solution([1,2,3,2,3])
```
