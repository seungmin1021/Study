```python
def solution(n):
    ans = 1
    a = ans
    num = [n]
    while n > 0:
        n = n // 2
        num.append(n)
    num.sort()

    for i in num[2:]:
        while True:
            a = a * 2
            if a == i:
                break
            else:
                a = a + 1
                ans += 1
                break
    return ans
```

```python
solution(5000)
```
