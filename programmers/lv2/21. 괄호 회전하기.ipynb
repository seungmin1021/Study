```python
def solution(s):
    answer = 0
    for i in range(len(s)):
        stack = []

        if s[0] == ']' or s[0] == ')' or s[0] == '}':
            s = s[1:] + s[0]
            continue
        for j in range(len(s)):
            if len(stack) == 0:
                stack.append(s[j])
            else:
                stack.append(s[j])
                if s[j] == ']':
                    if stack[-2] == '[':
                        stack.pop()
                        stack.pop()
                elif s[j] == ')':
                    if stack[-2] == '(':
                        stack.pop()
                        stack.pop()
                elif s[j] == '}':
                    if stack[-2] == '{':
                        stack.pop()
                        stack.pop()
        if len(stack) == 0:
            answer += 1
        s = s[1:] + s[0]
        print(s)
    return answer
```

```python
solution("[](){}")
```
