```python
def solution(cacheSize, cities):
    answer = 0
    cache = []
    if cacheSize == 0:
        return len(cities) * 5
    for i in cities:
        if i.lower() not in cache:
            if len(cache) < cacheSize:
                cache.append(i.lower())
                answer += 5
            else:
                cache.pop(0)
                cache.append(i.lower())
                answer += 5
        else:
            cache.pop(cache.index(i.lower()))
            cache.append(i.lower())
            answer += 1
            
    return answer
```

```python
solution(5,["Jeju", "Pangyo", "Seoul", "NewYork", "LA", "SanFrancisco", "Seoul", "Rome", "Paris", "Jeju", "NewYork", "Rome"])
```
