```python
def solution(s):
    stack = []
    a = list(s)
    for i in range(len(s)):
        if len(stack) == 0:
            stack.append(s[i])
            continue
        if s[i] == stack[-1]:
            stack.pop()
        else:
            stack.append(s[i])
    if len(stack) == 0:
        return 1
    else:
        return 0
```
