```python
import math
def solution(fees, records):
    answer = []
    num = []
    dic = {}
    a = []
    total = {}
    for i in records:
        t, n, io = i.split(' ')
        total[n] = 0
        num.append(n)
    num = sorted(set(num))
    for i in records:
        t, n, io = i.split(' ')
        if io == 'IN':
            h, m = t.split(':')
            dic[n] = int(h)*60 + int(m)
            a.append(n)
        else:
            h, m = t.split(':')
            total[n] += (int(h)*60 + int(m))-dic[n]
            a.pop(a.index(n))
    for i in a:
        total[i] += 1439 - dic[i]
    for i in num:
        if total[i] <= fees[0]:
            answer.append(fees[1])
        else:
            answer.append(fees[1] + math.ceil((total[i] - fees[0]) / fees[2]) * fees[3])
    return answer
```
