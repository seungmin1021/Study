```python
def solution(msg):
    answer = []
    dic = {chr(e + 64): e for e in range(1, 27)}
    msg = list(msg)
    cur = 27
    a = []
    while msg:
        s = ''
        s += msg.pop(0)
        for i in msg:
            s += i
            if s not in dic:
                answer.append(dic[s[:-1]])
                dic[s] = cur 
                cur += 1
                for i in range(len(s)-2):
                    msg.pop(0)
                break
            else:
                continue
        a.append(s)
    answer.append(dic[a[len(answer)]])
    return answer
```

```python
def solution(msg):
    answer = []
    dic = {chr(i + 64): i for i in range(1, 27)}
    cur = 27
    while msg:
        t = 1
        while msg[:t] in dic and t <= len(msg):
            t += 1
        t -= 1
        answer.append(dic[msg[:t]])
        dic[msg[:t+1]] = cur
        cur += 1
        msg = msg[t:]
    return answer
```
