```python
def solution(n,a,b):
    answer = 0

    for i in range(n):
        if a + 1 == b or b + 1 == a:
            if a + b % 4 == 3:
                return answer + 1
                
        if a % 2 == 0:
            a = int(a / 2)
        else:
            a = int(a // 2) + 1
        if b % 2 == 0:
            b = int(b / 2)
        else:
            b = int(b // 2) + 1
        answer += 1
```

```python
solution(8,4,5)
```
