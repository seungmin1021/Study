```python
def solution(p):
    answer = True
    p.sort()
    for i in range(len(p)-1):
        if p[i+1][:len(p[i])] == p[i]:
            answer = False
            return answer
    return answer
```

```python
def solution(phone_book):
    answer = True
    hash_map = {}
    for phone_number in phone_book:
        hash_map[phone_number] = 1

    for phone_number in phone_book:
        temp = ""
        for number in phone_number:
            temp += number
            if temp in hash_map and temp != phone_number:
                return False
    return answer
```

```python
solution(['5432', "1195524421", "119", "97674223"])
```
