```python
def solution(numbers, target):
    answer = 0
    def dfs(idx, result):
        nonlocal answer
        if idx == len(numbers):
            if result == target:
                answer += 1
        else:
            dfs(idx+1, result+numbers[idx])
            dfs(idx+1, result-numbers[idx])
    dfs(0,0)
    return answer
```

```python
solution([1,1,1,1,1], 3)
```

```python
from itertools import product
numbers = [1,1,1,1,1]
target = 3
l = [(x, -x) for x in numbers]
s = list(map(sum, product(*l)))
s.count(target)
```
