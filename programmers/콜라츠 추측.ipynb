```python
def solution(num):
    answer = 0
    cnt = 0
    while True:
        if num == 1:
            return cnt
        if num % 2 == 0:
            num = int(num / 2)
        elif num % 2 == 1:
            num = num * 3 + 1
        cnt += 1
        if cnt >= 500:
            return -1
    return answer
```
