```python
def solution(s):
    answer = ''
    a = list(s.split(' '))
    for i in range(len(a)):
        for j in range(len(a[i])):
            if j % 2 == 0 :
                answer += a[i][j].upper()
            else:
                answer += a[i][j].lower()
        if i < (len(a)-1):
            answer += ' '
        else:
            return answer
```
